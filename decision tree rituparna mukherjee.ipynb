```python
drug_df
```

```python
X = drug_df.iloc[:, :-1].values
```

```python
Y = drug_df.iloc[:,-1].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.2,random_state= 3)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy',random_state=1)
classifier.fit(X_train, Y_train)
```

```python
classifier
```

```python
predTree = classifier.predict(X_test)
```

```python
predTree
```

```python
print (predTree [0:5])
print (Y_test [0:5])
```

```python
from sklearn import metrics

print("DecisionTrees's Accuracy: ", metrics.accuracy_score(Y_test, predTree))
```

```python
from sklearn import metrics 
print("training set f1_score:",metrics.f1_score(Y_train,classifier.predict(X_train),average='micro'))
print("test set f1_score:",metrics.f1_score(Y_test,predTree,average='micro'))
print("test set mean accuracy:",classifier.score(X_test,Y_test))
```
